```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3, 4, 5], [1, 4, 7, 8, 6])
plt.plot([1, 2, 3, 4, 5], [1, 4, 7, 8, 6], "o")
plt.plot([1, 2, 3, 4, 5], [1, 3, 4, 5, 8], "--")
plt.plot([1, 2, 3, 4, 5], [1, 3, 4, 5, 8], "o")
plt.title("2D")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["line 1", "line 1, points ", "line 2", "line 2 points"])

plt.grid()

plt.show()
```
